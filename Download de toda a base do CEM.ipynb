# Fazendo Download de toda a base de dados do CEM disponível para downlaod

```python
import wget
import json
import os
```

```python
f =  open('resultados/inventario-download-de-dados-CEM.json', 'r')
```

```python
downloads = json.load(f)
```

```python
url_prefix = 'https://centrodametropole.fflch.usp.br'
```

```python
for i, l in enumerate(downloads):
    for d in l['Downloads']:
        filename = f'downloads/{str(i).zfill(3)}-{d["arquivo"]}'
        if os.path.exists(filename):
            print(f'{str(i).zfill(3)}-{d["arquivo"]} existe!')
        else:
            print(f'Downloading {filename} ...')
            try:
                wget.download(url_prefix + d['url'], out=f'downloads/{str(i).zfill(3)}-{d["arquivo"]}')
            except: 
                print(f'Problema com o dowwnload {filename} ...')
```
